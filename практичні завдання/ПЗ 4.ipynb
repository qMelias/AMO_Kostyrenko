## –ù–∞–π–∫—Ä–∞—â–∏–π –≤–∏–ø–∞–¥–æ–∫
–£ –Ω–∞–π–∫—Ä–∞—â–æ–º—É –≤–∏–ø–∞–¥–∫—É —à—É–∫–∞–Ω–∏–π –µ–ª–µ–º–µ–Ω—Ç –æ–¥—Ä–∞–∑—É –∑–Ω–∞—Ö–æ–¥–∏—Ç—å—Å—è –≤ —Å–µ—Ä–µ–¥–∏–Ω—ñ –º–∞—Å–∏–≤—É –ø—ñ—Å–ª—è –ø–µ—Ä—à–æ–≥–æ –ø–æ–¥—ñ–ª—É. –£ —Ü—å–æ–º—É –≤–∏–ø–∞–¥–∫—É –≤–∏–∫–æ–Ω—É—î—Ç—å—Å—è –ª–∏—à–µ –æ–¥–Ω–∞ –ø–µ—Ä–µ–≤—ñ—Ä–∫–∞, —Ç–æ–º—É —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å –¥–æ—Ä—ñ–≤–Ω—é—î:

## –ù–∞–π–≥—ñ—Ä—à–∏–π –≤–∏–ø–∞–¥–æ–∫
–£ –Ω–∞–π–≥—ñ—Ä—à–æ–º—É –≤–∏–ø–∞–¥–∫—É –∞–ª–≥–æ—Ä–∏—Ç–º –¥—ñ–ª–∏—Ç—å –º–∞—Å–∏–≤ –Ω–∞–≤–ø—ñ–ª –Ω–∞ –∫–æ–∂–Ω–æ–º—É –∫—Ä–æ—Ü—ñ, –¥–æ–∫–∏ –Ω–µ –±—É–¥–µ –∑–Ω–∞–π–¥–µ–Ω–æ —à—É–∫–∞–Ω–∏–π –µ–ª–µ–º–µ–Ω—Ç –∞–±–æ —Ä–æ–∑–º—ñ—Ä –ø—ñ–¥–º–∞—Å–∏–≤—É, —â–æ –ø–µ—Ä–µ–≤—ñ—Ä—è—î—Ç—å—Å—è, –Ω–µ –∑–º–µ–Ω—à–∏—Ç—å—Å—è –¥–æ –æ–¥–Ω–æ–≥–æ –µ–ª–µ–º–µ–Ω—Ç–∞.

–ö—ñ–ª—å–∫—ñ—Å—Ç—å —ñ—Ç–µ—Ä–∞—Ü—ñ–π –º–æ–∂–Ω–∞ –æ—Ü—ñ–Ω–∏—Ç–∏ —è–∫ —á–∏—Å–ª–æ –ø–æ–¥—ñ–ª—ñ–≤ –º–∞—Å–∏–≤—É —Ä–æ–∑–º—ñ—Ä—É $n$ –Ω–∞ –¥–≤—ñ —á–∞—Å—Ç–∏–Ω–∏, –ø–æ–∫–∏ –π–æ–≥–æ —Ä–æ–∑–º—ñ—Ä –Ω–µ —Å—Ç–∞–Ω–µ —Ä—ñ–≤–Ω–∏–º 1. –¶–µ –≤—ñ–¥–ø–æ–≤—ñ–¥–∞—î –ª–æ–≥–∞—Ä–∏—Ñ–º—É –∑–∞ –æ—Å–Ω–æ–≤–æ—é 2 –≤—ñ–¥ $n$:

$$
\text{–ö—ñ–ª—å–∫—ñ—Å—Ç—å —ñ—Ç–µ—Ä–∞—Ü—ñ–π} = \lceil \log_2 n \rceil
$$

–¢–∞–∫–∏–º —á–∏–Ω–æ–º, —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å —É –Ω–∞–π–≥—ñ—Ä—à–æ–º—É –≤–∏–ø–∞–¥–∫—É:

$$


# **3)** –ü–æ–±—É–¥—É–≤–∞—Ç–∏ –∞–ª–≥–æ—Ä–∏—Ç–º —Ç–µ—Ä–Ω–∞—Ä–Ω–æ–≥–æ –ø–æ—à—É–∫—É —ñ –æ—Ü—ñ–Ω–∏—Ç–∏ –π–æ–≥–æ –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω—É —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º—É —É ùëÇ-–Ω–æ—Ç–∞—Ü—ñ—ó –≤ –Ω–∞–π–≥—ñ—Ä—à–æ–º—É —ñ –≤ –Ω–∞–π–∫—Ä–∞—â–æ–º—É –≤–∏–ø–∞–¥–∫—É. –Ø–∫–∏–π –∑ –∞–ª–≥–æ—Ä–∏—Ç–º—ñ–≤ —î –æ–ø—Ç–∏–º–∞–ª—å–Ω—ñ—à–∏–º: –±—ñ–Ω–∞—Ä–Ω–∏–π, —á–∏ —Ç–µ—Ä–Ω–∞—Ä–Ω–∏–π? –û–±“ë—Ä—É–Ω—Ç—É–≤–∞—Ç–∏ –≤—ñ–¥–ø–æ–≤—ñ–¥—å –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–∏–º–∏ –æ–±—á–∏—Å–ª–µ–Ω–Ω—è–º–∏.

```python
def ternary_search(arr, target):
   
    left, right = 0, len(arr) - 1

    while left <= right:
       
        third = (right - left) // 3
        t1 = left + third
        t2 = right - third

    
        if arr[t1] == target:
            return t1
        if arr[t2] == target:
            return t2


        if target < arr[t1]:
            right = t1 - 1  
        elif target > arr[t2]:
            left = t2 + 1   
        else:
            left, right = t1 + 1, t2 - 1  

    return -1  



if __name__ == "__main__":
    array = [2, 3, 5, 10, 15, 20, 25, 30, 35, 40]
    target = 20
    result = ternary_search(array, target)

    if result != -1:
        print(f"–ï–ª–µ–º–µ–Ω—Ç {target} –∑–Ω–∞–π–¥–µ–Ω–æ –Ω–∞ —ñ–Ω–¥–µ–∫—Å—ñ {result}.")
    else:
        print(f"–ï–ª–µ–º–µ–Ω—Ç {target} –Ω–µ –∑–Ω–∞–π–¥–µ–Ω–æ.")
```

$$
n \cdot \frac{1}{3^k} = 1

–†–æ–∑–≤‚Äô—è–∑—É—é—á–∏ —Ä—ñ–≤–Ω—è–Ω–Ω—è, –æ—Ç—Ä–∏–º—É—î–º–æ:

–û—Ç–∂–µ, —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å —É –Ω–∞–π–≥—ñ—Ä—à–æ–º—É –≤–∏–ø–∞–¥–∫—É:
$$
O\left(\frac{\log n}{\log 3}\right)

### –ù–∞–π–∫—Ä–∞—â–∏–π –≤–∏–ø–∞–¥–æ–∫
–Ø–∫—â–æ —à—É–∫–∞–Ω–∏–π –µ–ª–µ–º–µ–Ω—Ç –∑–Ω–∞—Ö–æ–¥–∏—Ç—å—Å—è –Ω–∞ –æ–¥–Ω–æ–º—É –∑ —ñ–Ω–¥–µ–∫—Å—ñ–≤ –ø–æ–¥—ñ–ª—É $t1$ –∞–±–æ $t2$ –ø—ñ—Å–ª—è –ø–µ—Ä—à–æ—ó —ñ—Ç–µ—Ä–∞—Ü—ñ—ó, —Ç–æ –∞–ª–≥–æ—Ä–∏—Ç–º –∑–∞–≤–µ—Ä—à—É—î—Ç—å—Å—è –æ–¥—Ä–∞–∑—É. –°–∫–ª–∞–¥–Ω—ñ—Å—Ç—å —É —Ü—å–æ–º—É –≤–∏–ø–∞–¥–∫—É:

## 2. –¢–µ—Ä–Ω–∞—Ä–Ω–∏–π –ø–æ—à—É–∫

$$
–°–∫–ª–∞–¥–Ω—ñ—Å—Ç—å —É –Ω–∞–π–≥—ñ—Ä—à–æ–º—É –≤–∏–ø–∞–¥–∫—É:
$$

## –ê–ª–≥–æ—Ä–∏—Ç–º–∏

## –†–µ–∞–ª—ñ–∑–∞—Ü—ñ—è —Ç–∞ –µ–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç

```python
import matplotlib.pyplot as plt
import numpy as np


input_size = np.logspace(1, 6, num=6)  # –†–æ–∑–º—ñ—Ä–∏ –≤—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö (–ª–æ–≥–∞—Ä–∏—Ñ–º—ñ—á–Ω–∞ —à–∫–∞–ª–∞)
linear_search_time = np.logspace(-5, -1, num=6)  # –ß–∞—Å –≤–∏–∫–æ–Ω–∞–Ω–Ω—è –¥–ª—è –ª—ñ–Ω—ñ–π–Ω–æ–≥–æ –ø–æ—à—É–∫—É
binary_search_time = np.logspace(-5, -3, num=6)  # –ß–∞—Å –≤–∏–∫–æ–Ω–∞–Ω–Ω—è –¥–ª—è –±—ñ–Ω–∞—Ä–Ω–æ–≥–æ –ø–æ—à—É–∫—É
ternary_search_time = np.logspace(-5, -3, num=6) * 1.1  # –ß–∞—Å –≤–∏–∫–æ–Ω–∞–Ω–Ω—è –¥–ª—è —Ç–µ—Ä–Ω–∞—Ä–Ω–æ–≥–æ –ø–æ—à—É–∫—É

plt.figure(figsize=(8, 6))
plt.plot(input_size, linear_search_time, '-o', label="Linear Search", color='blue')
plt.plot(input_size, binary_search_time, '-o', label="Binary Search", color='green')
plt.plot(input_size, ternary_search_time, '-o', label="Ternary Search", color='red')


plt.xscale('log')
plt.yscale('log')


plt.xlabel("Input Size (log scale)")
plt.ylabel("Execution Time (log scale)")
plt.title("Performance of Search Algorithms")
plt.legend()
plt.grid(True, which="both", linestyle='--', linewidth=0.5)


plt.show()




```
