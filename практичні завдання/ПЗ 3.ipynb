# **1)** Вивчити самостійно і записати (будь-яким способом) алгоритм бульбашкового сортування. Оцінити асимптотику алгоритму сортування методом бульбашки в найгіршому і в найкращому випадку. Порівняти за цими показниками бульбашковий алгоритм з алгоритмом сортування вставлянням. Чому на практиці бульбашковий алгоритм виявляється менш ефективним у порівнянні з сортуванням методом зливанням?

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):  
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr


arr = [7, 3, 22, 9, 5, 4,]
sorted_arr = bubble_sort(arr)
print("Відсортований масив:", sorted_arr)
```

# Аналіз часової складності бульбашкового сортування

## 1. **Найгірший випадок**
- **Умови:** Масив відсортований у зворотному порядку.
- **Кількість порівнянь:** $ \frac{n(n-1)}{2} $.
- **Асимптотична складність:** $ O(n^2) $.

---

## 2. **Найкращий випадок**
- **Умови:** Масив уже відсортований.
- **Оптимізована версія:** Алгоритм завершується, якщо за прохід не відбулося жодного обміну.
  - **Кількість порівнянь:** $ O(n) $.
- **Неоптимізована версія:** Продовжує виконання незалежно від стану масиву.
  - **Кількість порівнянь:** $ O(n^2) $.

---

## 3. **Середній випадок**
- **Умови:** Елементи масиву розташовані у випадковому порядку.
- **Асимптотична складність:** $ O(n^2) $.

---

## Порівняння з алгоритмом сортування вставлянням:

### Алгоритм сортування вставлянням:
- **Принцип роботи:** Масив поділяється на дві частини: відсортовану та невідсортовану. Кожен елемент із невідсортованої частини вставляється в потрібну позицію в межах відсортованої частини.

### Асимптотика сортування вставлянням:
1. **Найгірший випадок:** $ O(n^2) $ (масив відсортований у зворотному порядку).
2. **Найкращий випадок:** $ O(n) $ (масив вже впорядкований).
3. **Середній випадок:** $ O(n^2) $.

### Порівняння:
- У найкращому випадку сортування вставлянням працює ефективніше ($ O(n) $) порівняно з бульбашковим алгоритмом без оптимізацій, який залишається $ O(n^2) $.
- Алгоритм вставляння виконує менше операцій порівняння та переміщення елементів.

---

## Чому бульбашковий алгоритм менш ефективний?

1. **Кількість операцій:**
   - У бульбашковому сортуванні виконується багато зайвих обмінів, навіть якщо масив майже впорядкований.
   - Сортування методом злиття скорочує кількість операцій завдяки стратегії "поділу і злиття".

2. **Асимптотична складність:**
   - Для бульбашкового алгоритму найгірший і середній випадки становлять $ O(n^2) $, тоді як для сортування злиттям — $ O(n \log n) $.

3. **Реальна продуктивність:**
   - На великих масивах сортування злиттям демонструє значно кращу ефективність завдяки логарифмічній структурі злиття підмасивів.
   - Бульбашковий метод переважно використовується в навчальних цілях або для дуже малих масивів.


# **2)** Оцінити асимптотичну складність алгоритму сортування зливанням, скориставшись основною теоремою рекурсії

## Оцінка асимптотичної складності алгоритму сортування злиттям

Алгоритм сортування злиттям ґрунтується на принципі **"розділяй і владарюй"**:

1. Масив поділяється на дві рівні частини.  
2. Кожна з частин сортується рекурсивно.  
3. Після сортування дві частини об'єднуються в один відсортований масив.

---

## Основи рекурсивного аналізу

Для оцінки часу роботи рекурсивних алгоритмів використовується основна теорема рекурсії. Вона дозволяє визначити асимптотичну оцінку розв'язків рекурентних рівнянь типу:

$T(n) = a \cdot T\left(\frac{n}{b}\right) + f(n)$,

де:  
- $a$ — кількість підзадач, на які розбивається задача;  
- $b$ — коефіцієнт зменшення розміру задачі на кожному етапі;  
- $f(n)$ — витрати на виконання роботи поза рекурсією (наприклад, злиття масивів).

---

## Рекурентне рівняння для сортування злиттям

Для алгоритму злиття:  
- Масив розбивається на дві підзадачі: $a = 2$,  
- Розмір кожної підзадачі зменшується вдвічі: $b = 2$,  
- Процес злиття займає $f(n) = O(n)$ часу.

Рекурентне рівняння має вигляд:

$T(n) = 2 \cdot T\left(\frac{n}{2}\right) + O(n)$.

---

## Розрахунок асимптотичної складності

1. **Обчислюємо логарифмічний показник**:  
   $ \log_b(a) = \log_2(2) = 1.$  

2. **Аналізуємо $f(n)$ у порівнянні з $n^{\log_b(a)}$:**  
   - $f(n) = O(n)$,  
   - $n^{\log_2(2)} = O(n)$.  

Оскільки $f(n)$ зростає з тією ж швидкістю, що й $n^{\log_b(a)}$, згідно з основною теоремою рекурсії маємо:

$T(n) = O(n \log n)$.

---

## Висновок

Асимптотична складність алгоритму сортування злиттям:  

$T(n) = O(n \log n)$.  

Це робить його значно ефективнішим, ніж бульбашкове сортування ($O(n^2)$), особливо для роботи з великими масивами.



# **3)** Вивчити і записати (будь-яким способом) самостійно алгоритм швидкого сортування. Оцінити асимптотичну складність алгоритму швидкого сортування, скориставшись основною теоремою рекурсії

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[0]  
    less = [x for x in arr[1:] if x <= pivot]  
    greater = [x for x in arr[1:] if x > pivot]  
    return quick_sort(less) + [pivot] + quick_sort(greater)


arr = [12, 22, 65, 10, 3, 8]
sorted_arr = quick_sort(arr)
print("Відсортований масив:", sorted_arr)
```

# Алгоритм швидкого сортування (Quick Sort)

## Алгоритм швидкого сортування

Алгоритм швидкого сортування використовує принцип **"розділяй і владарюй"**. Його основні етапи такі:

1. **Вибір опорного елемента:** Опорний елемент використовується для поділу масиву.  
2. **Розподіл елементів:** Масив ділиться на дві частини:  
   - Ліва частина — елементи, менші за опорний.  
   - Права частина — елементи, більші за опорний.  
3. **Рекурсивне сортування:** Алгоритм рекурсивно застосовується до обох частин масиву.

---

### Кроки алгоритму:
1. Вибирається опорний елемент (наприклад, перший, останній або випадковий елемент масиву).  
2. Масив ділиться на два підмасиви:
   - Елементи, менші за опорний, утворюють лівий підмасив.  
   - Елементи, більші за опорний, утворюють правий підмасив.  
3. Рекурсивне сортування повторюється для кожного з підмасивів.

---

## Асимптотична складність алгоритму швидкого сортування

Оскільки алгоритм має рекурсивну структуру, для оцінки його часу роботи використовується **основна теорема рекурсії**.

### Рекурентне рівняння:

Алгоритм поділяє масив на дві частини і сортує кожну рекурсивно. Витрати на поділ становлять $O(n)$, адже кожен елемент порівнюється з опорним.

Рекурентне рівняння виглядає так:

$$
T(n) = 2T\left(\frac{n}{2}\right) + O(n)
$$

де:  
- $a = 2$ — кількість підзадач (два підмасиви),  
- $b = 2$ — розмір кожного підмасиву зменшується вдвічі,  
- $f(n) = O(n)$ — витрати на розподіл елементів.

---

## Аналіз складності за основною теоремою рекурсії

1. **Обчислюємо логарифмічний показник:**  
   $$
   \log_b(a) = \log_2(2) = 1.
   $$

2. **Аналіз $f(n)$ щодо $n^{\log_b(a)}$:**  
   - $f(n) = O(n)$,  
   - $n^{\log_2(2)} = O(n)$.  

Оскільки $f(n)$ і $n^{\log_b(a)}$ зростають однаково, згідно з основною теоремою рекурсії:  

$$
T(n) = O(n \log n).
$$

---

## Висновок

Асимптотична складність алгоритму швидкого сортування:

$$
T(n) = O(n \log n).
$$

- **Середній і найкращий випадок:** Алгоритм демонструє високу ефективність ($O(n \log n)$), що робить його одним із найшвидших алгоритмів сортування для великих масивів.  
- **Найгірший випадок:** При неефективному виборі опорного елемента (наприклад, якщо масив уже впорядкований) складність може зрости до $O(n^2)$.  

Для уникнення найгіршого випадку рекомендується використовувати вдосконалені стратегії вибору опорного елемента, такі як "медіана трьох" або випадковий вибір.

## Алгоритми сортування з квадратичною складністю

До алгоритмів сортування, які мають квадратичну складність у найгіршому випадку ($O(n^2)$), належать:

1. **Бульбашкове сортування**  
2. **Сортування вставками**  
3. **Сортування вибором**  

---

### Чому квадратична складність є проблемою для великих даних?

Для великих обсягів даних (велике $n$) квадратична складність призводить до суттєвих затримок. Це пов’язано з тим, що для кожного елемента потрібно виконувати численні порівняння або переміщення, через що час роботи алгоритму зростає пропорційно квадрату кількості елементів.

#### Приклад:  
- Для масиву з 1000 елементів потрібно близько 1 мільйона операцій.  
- Для масиву з 10,000 елементів кількість операцій зростає до 100 мільйонів.  

Через це алгоритми з квадратичною складністю стають непридатними для роботи з великими наборами даних, оскільки вони вимагають занадто багато часу.

## Перевага сортування злиттям над сортуванням вставками для великих наборів даних

### 1. Асимптотична складність
- **Сортування злиттям** має складність $O(n \log n)$ навіть у найгіршому випадку, що робить його набагато ефективнішим для великих масивів даних.  
- **Сортування вставками** має складність $O(n^2)$ у найгіршому випадку, що суттєво уповільнює його роботу на великих наборах даних.

---

### 2. Масштабованість
- **Сортування злиттям** добре масштабується для великих обсягів даних завдяки логарифмічному зменшенню кількості операцій на кожному етапі.  
- **Сортування вставками** стає малопридатним для роботи з великими наборами, оскільки кількість порівнянь і переміщень елементів зростає квадратично зі збільшенням розміру масиву.

---

### 3. Постійна кількість операцій при злитті
- У **сортуванні злиттям** злиття підмасивів займає $O(n)$ часу незалежно від початкового порядку елементів.  
- У **сортуванні вставками** кількість операцій значно зростає при вставленні, особливо якщо елементи не впорядковані, що робить цей алгоритм менш ефективним для великих обсягів даних.

---

### Висновок
**Сортування злиттям** є значно ефективнішим і краще підходить для обробки великих масивів, ніж **сортування вставками**, завдяки своїй асимптотичній ефективності та здатності масштабуватися для великих наборів даних.

## Алгоритми сортування в стандартних бібліотеках мов програмування

### 1. **Python**
- **Алгоритм**: У Python використовується **Timsort** — гібридний алгоритм, який поєднує сортування злиттям і сортування вставками.  
- **Часова складність**: $O(n \log n)$ у найгіршому випадку.  
- **Особливості**: Timsort оптимізований для роботи з реальними даними, ефективно використовуючи попередньо впорядковані ділянки, що зменшує кількість порівнянь і переміщень.

---

### 2. **Java**
- **Алгоритм**:  
  - Для об'єктів (`Arrays.sort()`): **Merge Sort**.  
  - Для примітивних типів (`int[]`): **Dual-Pivot Quicksort**.  
- **Часова складність**:  
  - **Merge Sort**: $O(n \log n)$ у найгіршому випадку.  
  - **Quicksort**: $O(n \log n)$ у середньому, але $O(n^2)$ у найгіршому випадку.  
- **Особливості**:  
  - **Merge Sort** використовується для об'єктів через його стабільність.  
  - **Dual-Pivot Quicksort** забезпечує високу швидкість для примітивних типів завдяки оптимізації роботи з пам'яттю.

---

### 3. **C++**
- **Алгоритм**: У стандартній бібліотеці C++ (STL) використовується **IntroSort** — гібридний алгоритм, що поєднує **QuickSort**, **HeapSort** і **Insertion Sort**.  
- **Часова складність**:  
  - **QuickSort**: $O(n \log n)$ у середньому, але $O(n^2)$ у найгіршому випадку.  
  - **HeapSort**: $O(n \log n)$ у найгіршому випадку.  
  - **Insertion Sort**: використовується для малих підмасивів.  
- **Особливості**:  
  - **IntroSort** автоматично переключається між QuickSort і HeapSort залежно від глибини рекурсії, забезпечуючи стабільну продуктивність.

---

### Висновок
У стандартних бібліотеках мов програмування застосовуються гібридні алгоритми сортування, які комбінують переваги різних методів. Це дозволяє досягти високої ефективності, стабільності та надійної роботи з різноманітними наборами даних.



#### **5)** Яка різниця між алгоритмами сортування злиттям і швидким сортуванням? У яких випадках краще використовувати кожен з цих алгоритмів?

## Порівняння алгоритмів сортування злиттям та швидкого сортування

### 1. **Сортування злиттям (Merge Sort)**
- **Метод**: Розділяй і владарюй. Масив ділиться на дві частини, які сортуються рекурсивно і зливаються в один відсортований масив.  
- **Часова складність**: $O(n \log n)$ для найгіршого, середнього та найкращого випадків.  
- **Просторова складність**: $O(n)$ через використання додаткової пам'яті для злиття.  
- **Стабільність**: Алгоритм стабільний, тобто зберігає порядок однакових елементів.

---

### 2. **Швидке сортування (Quick Sort)**
- **Метод**: Також використовує підхід "розділяй і владарюй". Масив поділяється відносно опорного елемента (pivot), після чого лівий і правий підмасиви сортуються рекурсивно.  
- **Часова складність**:  
  - **Найгірший випадок**: $O(n^2)$ (за поганого вибору опорного елемента).  
  - **Середній і найкращий випадки**: $O(n \log n)$.  
- **Просторова складність**: $O(\log n)$ — використовує менше пам'яті порівняно із сортуванням злиттям.  
- **Стабільність**: Алгоритм нестабільний, тобто порядок однакових елементів може змінюватися.

---

## Вибір алгоритму залежно від задачі

### **Коли обирати сортування злиттям (Merge Sort):**
- Потрібна **стабільність**, наприклад, при сортуванні за кількома критеріями.  
- Гарантована **часова складність $O(n \log n)$** в найгіршому випадку важлива для надійності на великих наборах даних.  
- Оптимально для сортування **даних із зовнішньої пам’яті** або великих файлів на диску.  

### **Коли обирати швидке сортування (Quick Sort):**
- Пріоритетна **швидкість роботи** для середнього випадку через менші накладні витрати на пам'ять.  
- Стабільність не є важливою.  
- Для **невеликих або середніх наборів даних** швидке сортування часто ефективніше.  
- Можна оптимізувати вибір **опорного елемента**, щоб зменшити ризик найгіршого випадку.

#### **6)** Які фактори слід враховувати при виборі алгоритму сортування для конкретної задачі?

## Фактори для вибору алгоритму сортування

## Критерії вибору алгоритму сортування

При виборі алгоритму сортування для конкретної задачі варто враховувати кілька ключових факторів:

---

### 1. **Розмір вхідних даних**
- Для **малих наборів даних** добре підходять прості алгоритми, такі як **сортування вставками** або **сортування вибором**.  
- Для **великих наборів даних** краще використовувати алгоритми зі складністю $O(n \log n)$, наприклад, **швидке сортування** або **сортування злиттям**.

---

### 2. **Часова складність**
- Якщо важлива **максимальна продуктивність**, вибирайте алгоритми зі складністю $O(n \log n)$ у найгіршому випадку, наприклад, **Merge Sort** або **Timsort**.  
- Для допустимості рідкісних випадків погіршення продуктивності можна використовувати **Quick Sort**, який зазвичай швидший у середньому випадку.

---

### 3. **Просторова складність**
- Якщо є обмеження на використання пам'яті, перевагу слід надати алгоритмам із низькою просторовою складністю, наприклад, **Quick Sort** ($O(\log n)$).  
- Алгоритми, такі як **Merge Sort**, які потребують додаткової пам’яті ($O(n)$), не підходять у системах з дефіцитом ресурсів.

---

### 4. **Стабільність сортування**
- Для збереження **порядку однакових елементів** слід використовувати **стабільні алгоритми**, наприклад, **Merge Sort** або **Timsort**.  
- Якщо стабільність не критична, можна вибирати алгоритми, як-от **Quick Sort** або **Heap Sort**.

---

### 5. **Тип даних**
- Для **примітивних типів даних** (чисел) добре підходять **Quick Sort** або **Heap Sort**.  
- Для **об’єктів чи структур даних** краще використовувати **Timsort** (у Python) або **Merge Sort**, які забезпечують стабільність.

---

### 6. **Властивості вхідних даних**
- Якщо дані **майже відсортовані**, ефективно працюють алгоритми, такі як **Timsort** або **сортування вставками**.  
- У випадках, коли є багато **однакових елементів**, алгоритми, оптимізовані під такі випадки, наприклад, **Timsort**, забезпечать кращу продуктивність.

---

### 7. **Простота реалізації**
- Для швидкої та простої реалізації можна використовувати **Bubble Sort** або **Insertion Sort**, хоча вони менш ефективні для великих даних.  
- Складніші у реалізації алгоритми, як-от **Merge Sort** або **Quick Sort**, забезпечують кращу продуктивність і масштабованість.

---

### 8. **Паралельність**
- Для **паралельного виконання** алгоритми, як-от **Merge Sort**, добре підходять для розподілених систем.  
- **Quick Sort** також можна паралелізувати, але це складніше через його нерівномірну структуру розділення.
